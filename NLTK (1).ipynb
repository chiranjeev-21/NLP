```python
#Searching among text
#1. Concordance- give you the line that contains the given word in the text
#2. Similar-
#3. Common_contexts

# A concordance view shows us every occurrence of a given word, together with some context
text1.concordance('monstrous')
# We can also search for a multi-word phrase by passing a list of strings:
text1.concordance(["monstrous", "size"])
```

```python
# Each ConcordanceLine object contains the left, right side context of the query word, the line where query word occurs and the offset is the position of
# the start of the query word. The left_print and right_print is the words that should be printed in the concordance.
text1.concordance_list('monstrous')
```

```python
# Using concordance(token) gives you the context surrounding the argument token. It will show you the sentences where token appears.

# Using similar(token) returns a list of words that appear in the same context as token. In this case the the context
# is just the words directly on either side of token.

text2.similar('affection')
```

```python
text2.similar("love")
```

```python
# NLTK does provide an additional method called common_contexts that shows when the use of a list of words share the same surrounding words.
text2.common_contexts(['monstrous', 'very'])
```

```python
#Ngrams and collocation
#1. Ngrams- collection of n words (if word is taken as smallest unit)
from nltk import ngrams
sentence='I am Dr. shashank singh my specialization is AI, Machine Learning and NLP'
n=2
bigrams=ngrams(sentence.split(),n)
```

```python
for grams in bigrams:
    print(grams)
```

```python
#Collocation - most commonly used combination of words 
# Pointwise mutual information = PMI
# pmi(word1,word2) = log [probability(number of times both words appears in a document together)/probability(word1)*probability(word2)]
from nltk.collocations import *
bigram_measures=nltk.collocations.BigramAssocMeasures()
finder=BigramCollocationFinder.from_words(text1)
finder.nbest(bigram_measures.pmi,100)
```

```python
from nltk.collocations import *
trigram_measures=nltk.collocations.TrigramAssocMeasures()
finder=TrigramCollocationFinder.from_words(text1)
finder.nbest(trigram_measures.pmi,100)
```

```python
#Stylistics- to study systematic differences or analysis of words between different genres #study of the textual meaning
import nltk
from nltk import FreqDist
modals=['can', 'could', 'would', 'will', 'may', 'must'] #modal verbs
# to find out the number of times these words are present in different genres
category=brown.categories()
for cat in category:
    words= brown.words(categories=cat)
    fdist= FreqDist([w.lower() for w in words])
    for modal in modals:
        print (modal, fdist[modal], cat)
```
